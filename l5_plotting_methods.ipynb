# Plotting methods

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("P87-Section-2-Dataset-Basics-Resources\\P87-S2-Dataset-Basics-Resources\\heart.csv")
df.head(2)
```

```python
ncols = 2
nrows = 2

fig, ax = plt.subplots(nrows = nrows, ncols = ncols, figsize = (10, 5))
fig.tight_layout()
index = 0
cols = ["chol", "trestbps", "thalach", "oldpeak"]

for i in range(nrows):
    for j in range(ncols):
        ax[i, j].scatter(df["age"], df[cols[index]], alpha = 0.8)
        ax[i, j].set_xlabel("age")
        ax[i, j].set_ylabel(cols[index])
        index += 1
```

```python
#save the figure

save_fig = ax[0, 0].get_figure()
```

```python
#save image in png and pdf formats

save_fig.savefig("plots.png", bbox_inches = "tight", transparent=True)
save_fig.savefig("plots.pdf", bbox_inches = "tight")
```
