# Visualizing data

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Bar plot

df = pd.read_csv("P87-Section-2-Dataset-Basics-Resources\\P87-S2-Dataset-Basics-Resources\\heart.csv")
```

```python
df.head()
```

```python
chest_pain = df.groupby(by = "cp").median().reset_index()
chest_pain.head()
```

## Bar plots

```python
chest_pain.plot.bar(x="cp", y="age");
```

```python
chest_pain.plot.bar(x="cp");
```

```python
fig, ax = plt.subplots()

ax.bar(chest_pain["cp"], chest_pain["age"], label = "age", color = ["red", "blue", "orange", "violet"])
ax.set_xlabel("cp")
ax.legend();
```

```python
ax =sns.barplot("cp", "age", errcolor="k", capsize=0.2, data = df)
```

## Scatter plots

```python
df.plot.scatter("age", "trestbps");
```

```python
fig, ax = plt.subplots()

scatter = ax.scatter(df["age"], df["trestbps"], c = df["cp"], alpha = 0.8)
ax.set_xlabel("age")
ax.set_ylabel("trestbps")

#adding a legend to scatterplot
handles, labels = scatter.legend_elements()
ax.legend(handles, labels);
```

```python
sns.scatterplot("age", "trestbps", hue = "age", edgecolors = "none", data = df);
```

## Line plots

```python
ages = df.groupby("age").median().reset_index()
ages.head()
```

```python
ages.plot.line("age", ["chol", "trestbps"])
```

```python
fig, ax = plt.subplots()

plot = ax.plot(ages["age"], ages[["chol", "trestbps"]], lw = 1)
ax.set_xlabel("age")
ax.legend(labels = ["chol", "trestbps"]);
```

```python
sns.lineplot("age", "chol", data = ages)
sns.lineplot("age", "trestbps", data = ages);
```
